```python
from pycaret.utils import enable_colab
enable_colab()
```

```python
from pycaret.datasets import get_data
data = get_data('kiva')
```

```python
data.shape
```

```python
#randomly selecting 1000 items
data = data.sample(1000, random_state=786).reset_index(drop=True)
data.shape
```

```python
#importing nlp model
from pycaret.nlp import *
```

```python
#initializing the pycaret experiment
exp_nlp101 = setup(data = data, target = 'en', session_id = 123)
```

```python
#converting target variable 'en' into a list
text_list = list(data['en'])
type(text_list)
```

```python
#initializing the pycaret experiment with list input
exp_nlp101_list = setup(data = data, target = 'en', session_id = 123)
```

```python
lda = create_model('lda')
print(lda)
```

```python
#changing number of topics to 6
lda2 = create_model('lda', num_topics = 6, multi_core = True)
print(lda2)
```

```python
#assigning model to dataset
lda_results = assign_model(lda)
lda_results.head()
```

```python
#plotting top 100 bigrams from the dataset
plot_model(plot = 'bigram')
```

```python
#plotting the document distribution by topic
plot_model(lda, plot = 'topic_distribution')
```

```python
#plotting t-SNE a technique to reduce dimensionality for 2d/3d projections.
plot_model(lda, plot = 'tsne')
```

```python
#displays a user interface for all of the available plots for a given model
evaluate_model(lda)
```
